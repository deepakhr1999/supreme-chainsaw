```python
import json
import libsql_experimental as libsql
import os

with open("meals_2026.json", "r") as file:
    meals = json.load(file)

conn = libsql.connect("nutrition.db", sync_url=os.environ.get("TURSO_DATABASE_URL", "??"), auth_token=os.environ.get("TURSO_AUTH_TOKEN", "??"))

conn.executemany("""
    INSERT INTO meals (id, meal_type, calories, protein_g, carbs_g, fat_g, logged_at)
    VALUES (?, ?, ?, ?, ?, ?, ?)
""", [
    (
        m["id"],
        m["meal_type"],
        m["calories"],
        m["protein_g"],
        m["carbs_g"],
        m["fat_g"],
        m["logged_at"]
    )
    for m in meals
])

conn.commit()
print(f"Inserted {len(meals)} rows")
```
